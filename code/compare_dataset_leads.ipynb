```python
from datasets.deepfake_ecg.Deepfake_ECG_Dataset import Deepfake_ECG_Dataset, CH_8_2D_MATRIX_OUTPUT_TYPE, HR_PARAMETER
from datasets.PTB_XL_Plus.PTB_XL_PLUS_ECG_Dataset import PTB_XL_PLUS_ECGDataset, HR_PARAMETER, QRS_PARAMETER, PR_PARAMETER, QT_PARAMETER, SUB_DATASET_A, SUB_DATASET_B
import matplotlib.pyplot as plt
```

```python
data_ptbxl = PTB_XL_PLUS_ECGDataset(HR_PARAMETER, num_of_leads=8, sub_dataset="A", limit_dataset_for_testing=True)
```

```python
first_record_ptbxl = data_ptbxl[1][0]

import matplotlib.pyplot as plt
import numpy as np


# Create a figure and subplots
fig, axs = plt.subplots(8, 1, figsize=(15, 20), sharex=True)
fig.suptitle('ECG Channels', fontsize=16)

# Plot each channel
for i in range(8):
    axs[i].plot(np.arange(5000), first_record_ptbxl[i, :])
    axs[i].set_ylabel(f'Channel {i+1}')
    axs[i].grid(True)

# Set common x-label
axs[-1].set_xlabel('Sample')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()

```
